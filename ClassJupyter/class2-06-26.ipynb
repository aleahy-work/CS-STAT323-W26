## Make some data ##

```python
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
rng = np.random.default_rng()
```

```python
means1 = [.5, .5] # Mean of the distribution
cov1 = [[.7, 0], [0, .7]] 
N = 50
samples1 = rng.multivariate_normal(means1, cov1, size=N).round(decimals=2)
newdf1 = np.append(samples1, np.zeros([50,1]),axis=1)
```

```python
means2 = [3, 5] # Mean of the distribution
cov2 = [[.7, 0], [0, .7]] 
N = 50
samples2 = rng.multivariate_normal(means2, cov2, size=N).round(decimals=2)
newdf2 = np.append(samples2, np.ones([50,1]), axis=1)
mynewdf = np.append(newdf1, newdf2, axis=0)
```

```python
means3 = [-2, -3] # Mean of the distribution
cov3 = [[.7, 0], [0, .7]] 
N = 50
samples3 = rng.multivariate_normal(means3, cov3, size=N).round(decimals=2)
newdf3 = np.append(samples3, 2*np.ones([50,1]), axis=1)
mynewdf = np.append(mynewdf, newdf3, axis=0)
```

```python
means4 = [-3, 6] # Mean of the distribution
cov4 = [[.7, 0], [0, .7]] 
N = 50
samples4 = rng.multivariate_normal(means4, cov4, size=N).round(decimals=2)
newdf4 = np.append(samples4, 3*np.ones([50,1]), axis=1)
mynewdf = np.append(mynewdf, newdf4, axis=0)
```

```python
thenewdf = pd.DataFrame({ 'myx': mynewdf[:,0], 'myy': mynewdf[:,1], 'myclass': mynewdf[:,2]})
```

```python
sns.scatterplot(thenewdf, x='myx', y='myy', hue='myclass')
```
