```python
import pyspark
pyspark.__version__
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("MyDemo") \
    .getOrCreate()
```

```python
df = spark.read.parquet(
    "hdfs://pi1.knoxds.org:8020/datasets/yellow_tripdata_2025-01.parquet"
)
```

```python
df.columns
```

```python
df.count()
```

```python
df.filter(df.passenger_count > 4).count()
```

```python
mydf = df.filter(df.passenger_count > 4).select('passenger_count')
```

```python
mydf.count()
```

```python
mydf.columns
```

```python
import pyspark.sql.functions as sf
mydf.select(sf.avg('passenger_count')).show()
```
