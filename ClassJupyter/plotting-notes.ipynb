# Matplotlib Quickstart #

As with pandas, the [Matlotlib API Reference Page](https://matplotlib.org/stable/api/index.html) is pretty awesome.  Click on the link and check it out as needed. 

```python
import matplotlib.pyplot as plt  # don't worry about the error message
```

## Line Plots ##

When combined with more x-values, this is useful for plotting graphs of functions

```python
myx = [1,3,5]
myy = [1,4,3]
```

```python
plt.plot(myx, myy)
plt.show()
```

## Saving a figure to a file ##

```python
plt.plot(myx, myy)
plt.savefig('myplot.png',format='png')  # You must put this before plt.show()
plt.show()
```

## Scatter Plots ##

```python
newx = [1,2,1,4,2,3,4,5]
newy = [1,5,3,4,2,3,4,5]
```

```python
plt.scatter(newx, newy)
plt.xlabel('My x label')  ## These next three lines add labels
plt.ylabel('My y label')
plt.title('My title')
plt.show()
```

## Layering Plots ##

```python
plt.plot(myx, myy)
plt.scatter(newx,newy)  # Just stack several things before the show()
plt.show()
```

## Categorical Data: Bar and Pie Charts ##

The hard part here is that the data has to *already be in a frequncy table before being used in a bar or pie chart*, so we do that first with Pandas.  There are probably easier ways than what I show.

```python
# This just creates a random data set of categorical variables and is 
# not necessary for what follows
import numpy as np
mydata = np.random.choice(['A','B','C'],100, [0.6,0.2,0.2])
mydata
```

```python
import pandas as pd
```

```python
mydf = pd.DataFrame({ 'Mydata': mydata})  # This creates as pandas dataframe for the data
```

```python
mydf.head(5)
```

```python
#  This creates a data frame with the labels in the first column
#  and the count in the second column.
#
myfreqtable = mydf.value_counts().to_frame().reset_index()  # This creates the frequency table
```

```python
myfreqtable
```

```python
plt.bar(myfreqtable['Mydata'], myfreqtable['count'])
plt.show()
```

```python
plt.pie(myfreqtable['count'], labels=myfreqtable['Mydata'])
plt.show()
```

## Histograms and Box-Wiskers Charts #

```python
mydata = np.random.normal(loc=70, scale=2.5, size=100)
```

```python
mydata  # This is a random sample from a normal distribution
# You don't need to do this.
```

```python
plt.hist(mydata)
plt.xlabel('My normal distribution')
plt.show()
```

```python
plt.boxplot(mydata)
plt.show()
```
