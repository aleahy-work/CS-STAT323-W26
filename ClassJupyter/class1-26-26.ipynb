The breast cancer dataset is described [here](https://archive.ics.uci.edu/dataset/17/breast+cancer+wisconsin+diagnostic).


```python
import pandas as pd
```

```python
mydata = pd.read_csv("Data/breast-cancer.csv")
```

```python
mydata.head(5)
```

```python
mydata.size
```

```python
mydata.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
myy = mydata["diagnosis"]
```

```python
myx = mydata.drop(['id','diagnosis'], axis=1)
```

```python
myx.head(5)
```

```python
X_train, X_test, y_train, y_test = train_test_split(myx, myy, test_size=0.2)
```

```python
X_train.head(5)
```

```python
from sklearn.tree import DecisionTreeClassifier
mydt = DecisionTreeClassifier()
```

```python
mydt
```

## First Try ##

```python
mydt.fit(X_train, y_train)
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(12, 8))
plot_tree(mydt) 
plt.show()
```

```python
yhat = mydt.predict(X_test)
```

```python
mydt.score(X_test, y_test)
```

## Second Try ##

```python
mydt2 = DecisionTreeClassifier(max_depth=4)
```

```python
mydt2.fit(X_train, y_train)
```

```python
plt.figure(figsize=(12, 8))
plot_tree(mydt2) 
plt.show()
```

```python
mydt2.score(X_test, y_test)
```

## Third Try ##

```python
mydt3 = DecisionTreeClassifier(max_depth=4, min_samples_leaf=5)
```

```python
mydt3.fit(X_train, y_train)
```

```python
plt.figure(figsize=(12, 8))
plot_tree(mydt3) 
plt.show()
```

```python
mydt3.score(X_test, y_test)
```

## Bagging ##

```python
from sklearn.ensemble import BaggingClassifier
```

```python
mybagged = BaggingClassifier().fit(X_train, y_train)
```

```python
mybagged
```

```python
mybagged.score(X_test, y_test)
```

```python
mybagged2 = BaggingClassifier(n_estimators=100).fit(X_train, y_train)
```

```python
mybagged2.score(X_test, y_test)
```
